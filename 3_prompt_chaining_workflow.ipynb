```python
from langgraph.graph import StateGraph , START, END
from typing import TypedDict
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
class BlogState(TypedDict):
    topic: str
    # keywords: list[str]
    outline: str
    blog_post: str
```

```python
def generate_outline(state : BlogState) -> BlogState:
    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.7)
    outline_prompt = f"""
    Generate a detailed outline for a blog post about the following topic: {state['topic']}
    """
    state['outline'] = llm.invoke(outline_prompt)
    return state
```

```python
def generate_blog_post(state : BlogState) -> BlogState:
    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.7)
    blog_post_prompt = f"""
    Write a detailed blog post based on the following outline: {state['outline']}
    """
    state['blog_post'] = llm.invoke(blog_post_prompt)
    return state
```

```python
graph = StateGraph(BlogState)

# add nodes
graph.add_node("generate_outline", generate_outline)
graph.add_node("generate_blog_post", generate_blog_post)

# add edges
graph.add_edge(START, "generate_outline")
graph.add_edge("generate_outline", "generate_blog_post")
graph.add_edge("generate_blog_post", END)

#compile the graph
workflow = graph.compile()

```

```python
# execute the workflow
initial_state = {'topic' : "The impact of AI on modern software development"}
final_state = workflow.invoke(initial_state)
print(final_state['blog_post'])
```
